```python
import numpy as np
import pandas as pd
import os
import matplotlib
import matplotlib.pyplot as plt
from joblib import load
```

## fixcost sensitivity

```python
df_sp = pd.read_excel('impact_fixcost.xlsx', sheet_name='SP')
df_wass = pd.read_excel('impact_fixcost.xlsx', sheet_name='Wass')
df_mad = pd.read_excel('impact_fixcost.xlsx', sheet_name='MAD')
```

```python
df_sp.head()
```

```python
x = np.arange(200, 501, 50)
df_sp['fixcost'] = np.repeat(x, 5)
df_wass['fixcost'] = np.repeat(x, 5)
df_mad['fixcost'] = np.repeat(x, 5)
```

```python
df_sp.head()
```

```python
sp = df_sp.groupby('fixcost')[[ 'open_num', 'cost1','cost2', 'obj']].mean()
wass = df_wass.groupby('fixcost')[[ 'open_num', 'cost1','cost2', 'obj']].mean()
mad = df_mad.groupby('fixcost')[[ 'open_num','cost1', 'cost2', 'obj']].mean()
```

```python
sp['95th'] = df_sp.groupby('fixcost')['obj'].quantile(0.95).values
wass['95th'] = df_wass.groupby('fixcost')['obj'].quantile(0.95).values
mad['95th'] = df_mad.groupby('fixcost')['obj'].quantile(0.95).values
```

```python
with pd.ExcelWriter('impact_fixcost_分析.xlsx', engine='openpyxl') as writer:
    # 将每个DataFrame写入不同的工作表
    sp.to_excel(writer, sheet_name='SP', index=False)
    wass.to_excel(writer, sheet_name='Wass', index=False)
    mad.to_excel(writer, sheet_name='MAD', index=False)
```

## capacity sensitivity

```python
df_sp = pd.read_excel('impact_capacity.xlsx', sheet_name='SP')
df_wass = pd.read_excel('impact_capacity.xlsx', sheet_name='Wass')
df_mad = pd.read_excel('impact_capacity.xlsx', sheet_name='MAD')
```

```python
x = np.arange(30, 61, 5)
df_sp['cap'] = np.repeat(x, 5)
df_wass['cap'] = np.repeat(x, 5)
df_mad['cap'] = np.repeat(x, 5)
```

```python
sp = df_sp.groupby('cap')[[ 'open_num', 'cost1','cost2', 'obj']].mean()
wass = df_wass.groupby('cap')[[ 'open_num', 'cost1','cost2', 'obj']].mean()
mad = df_mad.groupby('cap')[[ 'open_num','cost1', 'cost2', 'obj']].mean()
```

```python
sp['95th'] = df_sp.groupby('cap')['obj'].quantile(0.95).values
wass['95th'] = df_wass.groupby('cap')['obj'].quantile(0.95).values
mad['95th'] = df_mad.groupby('cap')['obj'].quantile(0.95).values
```

```python
sp.reset_index(inplace=True)
wass.resert_index(inplace=True)
mad.reset_index(inplace=True)
sp
```

## radius sensitivity

```python
from joblib import load
```

```python
radius_impact = load('impact_radius.joblib')
```

```python
columns = ['theta', 'sd', 'y', 'open_num', 'cots1', 'cost2', 'costt']
df = pd.DataFrame(radius_impact, columns=columns)
```

```python
I = [6,9,12]
I = np.repeat(I, 55)
df['I'] = I
```

```python
df
```

```python
wass = df.groupby(['I', 'theta'])[[ 'open_num', 'cots1','cost2', 'costt']].mean()
wass.reset_index(inplace=True)
wass
```

```python
quantile95 = df.groupby(['I', 'theta'])['costt'].quantile(0.95).values
wass['95th'] = quantile95
```

```python
wass.to_csv('impact_radius_分析.csv', index=False)
```
