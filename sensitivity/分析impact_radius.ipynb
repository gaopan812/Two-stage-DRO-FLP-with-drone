```python
import numpy as np
import pandas as pd
import os
import matplotlib
import matplotlib.pyplot as plt
from joblib import load
```

```python
files = os.listdir('.')
print(files)
```

```python
radius_impact = load('impact_radius.joblib')
```

```python
radius_impact[:5]
```

```python
df = pd.DataFrame(radius_impact[:5])
```

```python
columns = ['theta', 'sd', 'y', 'open_num', 'cots1', 'cost2', 'costt']
df = pd.DataFrame(radius_impact, columns=columns)
df
```

```python
I = [6,9,12]
I = np.repeat(I, 55)
df['I'] = I
```

```python
df
```

```python
quantile95 = df.groupby(['I', 'theta'])['costt'].quantile(0.95)
quantile95
```

```python
recourse = df.groupby(['I', 'theta'])['cost2'].mean()
recourse
```

```python
meantotal = df.groupby(['I', 'theta'])['costt'].mean()
meantotal
```

## recourse

```python
# 统一设置字体
plt.rcParams["font.family"] = 'Times New Roman'
 
# 分别设置mathtext公式的正体和斜体字体
matplotlib.rcParams['mathtext.fontset'] = 'custom'
matplotlib.rcParams['mathtext.rm'] = 'Times New Roman'  # 用于正常数学文本
matplotlib.rcParams['mathtext.it'] = 'Times New Roman:italic'  # 用于斜体数学文本
plt.rcParams.update({'font.size': 15})

wass_6 = recourse[:11].values
wass_9 = recourse[11:22].values
wass_12 = recourse[22:33].values
x = np.arange(0.5, 8.1, 0.75)
# 创建一个标准的图形窗口,并设置其尺寸
fig = plt.figure(figsize=(8, 6))

plt.grid()
plt.plot(x, wass_6, marker = '*', label='(I,J,L)=(6,3,3)')
plt.plot(x, wass_9, marker = '^', label='(I,J,L)=(9,4,4)')
plt.plot(x, wass_12, marker = 'v', label='(I,J,L)=(12,5,5)')
plt.legend(loc='best')
plt.xlabel(r'Wasserstein radius $\theta$', fontsize=18)
plt.ylabel('Recourse cost', fontsize=18)
plt.xticks(x)
# 设置x、y轴刻度字体大小
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)

plt.savefig('figures/radius_recourse.pdf', dpi=300)

#plt.show()
```

## ratio

```python
# 统一设置字体
plt.rcParams["font.family"] = 'Times New Roman'
 
# 分别设置mathtext公式的正体和斜体字体
matplotlib.rcParams['mathtext.fontset'] = 'custom'
matplotlib.rcParams['mathtext.rm'] = 'Times New Roman'  # 用于正常数学文本
matplotlib.rcParams['mathtext.it'] = 'Times New Roman:italic'  # 用于斜体数学文本
plt.rcParams.update({'font.size': 14})

wass_6 = 1311.076/(quantile95[:11].values)
wass_9 = 2123.737/(quantile95[11:22].values)
wass_12 = 2348.089/(quantile95[22:33].values)
x = np.arange(0.5, 8.1, 0.75)
# 创建一个标准的图形窗口,并设置其尺寸
fig = plt.figure(figsize=(8, 6))
ax=plt.gca()
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['left'].set_linewidth(1.5)
ax.spines['right'].set_linewidth(1.5)
ax.spines['top'].set_linewidth(1.5)
# 在Y=1的位置绘制一条水平线
plt.axhline(y=1, color='black', linestyle='-', linewidth=2)  # 红色虚线

plt.grid(linestyle='-.')
plt.plot(x, wass_6, marker = 's', markersize = 10, linewidth=2,label='(I,J,L)=(6,3,3)')
plt.plot(x, wass_9, marker = '^', markersize = 10, linewidth=2,label='(I,J,L)=(9,4,4)')
plt.plot(x, wass_12, marker = 'v', markersize = 10, linewidth=2,label='(I,J,L)=(12,5,5)')
plt.legend(loc='best')
plt.xlabel(r'Wasserstein radius $\theta$', fontsize=18)
plt.ylabel('Ratio', fontsize=18)
# 设置x、y轴刻度字体大小
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.savefig('figures/radius_ratio_total.pdf', dpi=300)
```
